### Model Download

```python
from unsloth import FastLanguageModel
import torch

model_name = "mistralai/Mistral-7B-Instruct-v0.2" 
max_seq_length = 4*1024
dtype = None 
load_in_4bit = False 

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    token = hf_token,
)
```

### Quantized GGUF Upload

```python

model_name = "mistral_7b_instruct"
quantization_methods = ["f16", "q8_0"]

# save multiple variants
if True:
    for quantization_method in quantization_methods:
        full_model_name = f"{model_name}_{quantization_method}"
        model.push_to_hub_gguf(f"JanJacobsen/{full_model_name}", tokenizer, quantization_method=quantization_method, token=hf_token)

# save locally
if False:
    model.save_pretrained_gguf(full_model_name, tokenizer, quantization_method=quantization_method)
```
